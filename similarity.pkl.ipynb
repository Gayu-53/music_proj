```python
new_df=df[['Song-Name','tags']]
new_df
```

```python
has_floats = new_df['tags'].dtype == 'float64' 
if has_floats:
    print("There are float values in the 'tags' column. Please investigate and clean them.")
df
```

```python
new_df = new_df[new_df['tags'].notna()]
```

```python
new_df['tags'] = new_df['tags'].astype(str)
new_df
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
new_df
```

```python
from sklearn.feature_extraction.text import CountVectorizer # type: ignore
cv=CountVectorizer(max_features=2000)
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors.shape
```

```python
cv.get_feature_names_out() 
```

```python
from sklearn.metrics.pairwise import cosine_similarity # type: ignore
similarity=cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])
```

```python
new_df.rename(columns={'Song-Name':'title'},inplace=True)
```

```python
def recommend(music):
    music_index=new_df[new_df['title']==music].index[0]
    distances=similarity[music_index]
    music_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    for i in music_list:
        print(new_df.iloc[i[0]].title) 
              
```

```python
recommend('O Sathiya')
```

```python
recommend('Jane Kahan Gayi')
```
